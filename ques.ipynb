```python
import matplotlib.pyplot as plt
x = [2,4,6,8,10]
y = [20,60,40,61,10]
plt.title("graph")
plt.xlabel("std. id")
plt.ylabel("std. marks")
plt.plot(x, y, color='blue', marker='.', linestyle='--', linewidth=1, markersize=15)
plt.text(
    8,
    61,
    "Peak"
)
plt.annotate(
    "lowest Point",
    xy=(10,10),
    xytext=(10,20),
    arrowprops=dict(
        arrowstyle='->'
    )
)
plt.show()


```

```python

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title("graph of trignometri funtions", color='red')
plt.xlabel("angle(in radians)")
plt.ylabel("value of function")
# Fix y-axis range so 0 is in the middle
plt.ylim(-1.5, 1.5)
plt.grid(visible=True, color="brown", linestyle='--')
plt.plot(x, y1, linewidth=3,color = "blue", label="sin(x)")
plt.plot(x, y2, linewidth=3 ,color= "red", label="cos(x)")
plt.xticks(
    [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
    ['0', 'π/2', 'π', '3π/2', '2π']
)
plt.legend()
# Draw x-axis at y=0
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.savefig("trignometry.png",dpi = 300)

plt.show()
```

```python
# improved version
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8,5))

plt.title("Graph of Trigonometric Functions", color="red")
plt.xlabel("Angle (in radians)")
plt.ylabel("Value of Function")

plt.ylim(-1.5, 1.5)

plt.grid(
    visible=True,
    color="brown",
    linestyle="--",
    alpha=0.5
)

plt.plot(
    x,
    y1,
    linewidth=2,
    color="blue",
    label="sin(x)"
)

plt.plot(
    x,
    y2,
    linewidth=2,
    color="red",
    label="cos(x)"
)

plt.xticks(
    [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
    ['0', 'π/2', 'π', '3π/2', '2π']
)

plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.legend()

plt.savefig("trigonometry.png", dpi=300)

plt.show()
```
